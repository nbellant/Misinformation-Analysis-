```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
import json
```

```python
#Open and Load the JSON file
with open('covid19_twitter_misinf_dataset.json') as f:
  data = pd.read_json(f, lines=True, orient='columns')

print(data)
```

```python
#First things first, I need to create lists to store my real/fake information, keep my counts, and also apply filters for which to identify my information.

# List to store entries for misinformation
mis_info = []

# Creates a list to store my entries for true/real
true_info = []

# Filters to identify the information category
true_cat = ['bot_flag\":\"TRUE']
mis_cat = ['bot_flag\":\"FALSE']

# These counters will be used to keep track of the total number of true or misinformation entries
trueinfo_count = 0
misinfo_count = 0
```

```python
# These are my specific hypothesis counters

#Counters to keep track of the instances of the referencing of naming of China 
china_true_count = 0
china_misinfo_count = 0

# Counters to track instances of the Presidental name substrings (Joe and Trump)
prez_true_count = 0
prez_misinfo_count= 0

# Counters to track instances of the substring "!"" in the tweets
exc_true_count = 0
exc_misinfo_count = 0
```

```python
# I want to sort through each entry in the JSON file to group it in the True info section
for d in data:
    
  # Check if the tweet falls in the true category
  if true_cat[0] in d:
    true_info.append(d)
    trueinfo_count = trueinfo_count + 1

    # Next, I take the True info and look through every word in the entry and count instances of my different hypotheses
    for w in d.split():
      if 'China' in w.lower():
        china_true_count = china_true_count + 1
      if 'Trump' in w.lower() or 'Joe' in w.lower():
        prez_true_count = prez_true_count + 1
      if '!' in w.lower():
        exc_true_count = exc_true_count + 1

  # I want to sort through each entry in the JSON file to group it in the False info section     
  elif mis_cat[0] in d:
    mis_info.append(d)
    misinfo_count = misinfo_count + 1

    # Next, I take the False info and look through every word in the entry and count instances of my different hypotheses
    for w in d.split():
      if 'China' in w.lower():
        china_misinfo_count = china_misinfo_count + 1
      if 'Trump' in w.lower() or 'Joe' in w.lower():
        prez_misinfo_count = prez_misinfo_count + 1
      if '!' in w.lower():
        exc_misinfo_count = exc_misinfo_count + 1
```

```python
# Mean for instances of the referencing of naming of China 
china_true_mean = china_true_count / trueinfo_count
china_misinfo_mean = china_misinfo_count / misinfo_count

# Mean for instances of the Presidental names (Joe and Trump)
prez_true_mean = prez_true_count / trueinfo_count
prez_misinfo_mean = prez_misinfo_count / misinfo_count

# Mean for instances of "!" in the tweets
exc_true_mean = exc_true_count / trueinfo_count
exc_misinfo_mean = exc_misinfo_count / misinfo_count

#Finally, I print all the answers cleanly so that I can better compare the results
print("China/True information is: " + str(china_true_mean))
print("China/Misinformation is: " + str(china_misinfo_mean))
print("Prez/True information is: " + str(prez_true_mean))
print("Prez/Misinformation is: " + str(prez_misinfo_mean))
print("Exc/True information is: " + str(exc_true_mean))
print("Exc/Misinformation is: " + str(exc_misinfo_mean))
```
